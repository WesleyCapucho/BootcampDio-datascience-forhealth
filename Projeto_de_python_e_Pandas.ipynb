```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
#Criando um dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Vizualizando as 5 primeiras linhas
df.head()
```

```python
#Quantidade de linhas por coluna
df.shape
```

```python
#Verificando os tipos de dados
df.dtypes
```

```python
#Encontrando a reita total
df["Valor Venda"].sum()
```

```python
#Encontrando o custo total
df["Custo"] = df["Custo Unitário"]. mul(df["Quantidade"]) 
```

```python
df.head()
```

```python
#custo total
round(df["Custo"].sum(), 2)
```

```python
df ["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head()
```

```python
#Lucro Total
round(df["Lucro"].sum(),2)
```

```python
#Criando uma coluna com o total de dias de envio
df["Tempo de Envio"] = df["Data Envio"] - df ["Data Venda"]
```

```python
df.head()
```

**Tempo médio de envio para cada produto**

```python
#Extraind apenas os dias
df["Tempo de Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head()
```

```python
#Verificando o tipo da coluno Tempo de Envio
df["Tempo de Envio"].dtype
```

```python
#Média do tempo de envio por marca
df.groupby("Marca")["Tempo de Envio"].mean()
```

**Valores Perdidos**

```python
#verificando a existência de valores faltantes
df.isnull().sum()
```

**Lucro por ano e marca**

```python
#Agrupando por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"]) ["Lucro"].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#Alterando o index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index() 

lucro_ano
```

```python
#Total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto") ["Quantidade"].sum().sort_values(ascending= True).plot.bar(title = "Total de Produtos VEndidos")
plt.xlabel("total")
plt.ylabel ("Produtos");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.barh(title= "Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel ("Receita")
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando as vendas do ano de 2008
df_2008 = df[df["Data Venda"].dt.year == 2008]
```

```python
df.head()
```

```python
df_2008.groupby(df_2008["Data Venda"].dt.month) ["Lucro"].sum().plot(title = "Lucro x Mês")
plt.xlabel ("Mês")
plt.ylabel ("Lucro");
```

```python
df_2008.groupby("Marca") ["Lucro"].sum().plot.bar(title = "Lucro x Marca")
plt.xlabel ("Marca")
plt.ylabel ("Lucro")
plt.xticks(rotation = "horizontal");
```

```python
df_2008.groupby("Classe") ["Lucro"].sum().plot.barh(title = "Lucro x classe")
plt.xlabel ("Classe")
plt.ylabel ("Lucro")
plt.xticks(rotation = "horizontal");
```
